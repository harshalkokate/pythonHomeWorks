```python
def main():
    # Call to handler function
    handler_function()

# Goes to 'value_check_and_menu_function' if the value entered is between 1 to 4, Otherwise exits on 5. 
def handler_function():
    while True:
        value_for_task = value_check_and_menu_function()
        if value_for_task != 5:
            func = switcher.get(value_for_task, "Invalid")
            func()
            continue
        else:
            print()
            print('Good bye..!')
            print('Have a good one..!')
            break
            
# Displays the created menus and checks if the entered option is valid or not.            
def value_check_and_menu_function():
    while True:
        print()
        print('''Bank account application
            1) Create new account
            2) Credit/Debit an account 
            3) List all accounts
            4) List account history
            5) Quit''')
        try:
            print()
            value = int(input('What would you like to do: '))
            print()
        except ValueError:
            print()
            print('Please enter the valid option.')
            print()
            continue

        if not (isinstance(value,int) and 1<=value<=5):
            print("Please enter a valid option.")
            continue
        else:
            break
    return value

# Creates account for individuals. 
def create_account():
    print('Creating a new account...')
    print('Please enter the individuals')
    while True:
        first_name = input('First name: ')
        last_name = input('Last name: ')
        
        try:
            beg_bal = float(input('Beginning Balance (USD): '))
        except ValueError:
            print('Enter number in balance.')
            continue

        if type(first_name)==str and type(last_name)==str and type(beg_bal)==float:
            if len(account_list)!=0:
                account_list.append([account_list[-1][0] + 1,first_name,last_name,beg_bal,beg_bal])
                print('')
                print('New account created for {} {} (Account# {})'.format(first_name,last_name,account_list[-1][0])) 
            else:
                account_list.append([1001,first_name,last_name,beg_bal,beg_bal])
                print('')
                print('New account created for {} {} (Account# {})'.format(first_name,last_name,1001))
            break
        else:
            print('Either First Name or Last name is not word.')
            print()
            continue

# Credit and debit the accounts for individual account number.
def crediting_and_debiting():
    print('Crediting/Debiting an account...')
    acc_num = 0
    while True:
        try:
            acc_num = int(input('Please enter account number: '))
        except ValueError:
            print('Enter valid account number.')
            continue
            
        count = 0   
        for i in account_list:
                if i[0]== acc_num:
                    count += 1
                   
        if acc_num > 1000 and count==1:
            while True:
                try:
                    amount = float(input('Please enter the amount: '))
                except ValueError:
                    print('Enter a valid amount.')

                if amount < 0: 
                    t_list = transaction_handler(amount,acc_num)
                    print()
                    if type(t_list) == str:
                        print(t_list)
                    else:    
                        print('{} {} (Account #{}) debited ${}'.format(t_list[1],t_list[2],t_list[0],str(amount)[1:]))
                        print('New balance: ${}'.format(t_list[3]))        
                else:
                    t_list = transaction_handler(amount,acc_num)
                    print()
                    print('{} {} (Account #{}) credited ${}'.format(t_list[1],t_list[2],t_list[0],str(amount)))
                    print('New balance: ${}'.format(t_list[3]))
                break
            break
        else:
            print("Account Number doesn't exist.")
            break

# Handles the transactions of each credit and debit. 
def transaction_handler(amount,acc_num):
    for i in account_list:
        if i[0] == acc_num:
            if i[3] < (amount * -1):
                return 'Insufficient balance.'
            else:
                i.append(amount)
                i.append(i[3] + amount)
                i[3] += amount
                return i

# Display all accounts.
def list_account():
    if account_list:
        for i in account_list:
            print('{}  {} {}   ${:.2f}'.format(i[0],i[1],i[2],i[3]))
    else:
        print('No account yet')

# Display the transction history for all accounts.
def transaction_history():
    acc_num = 0
    while True:
        try:
            acc_num = int(input('Please enter account number: '))
        except ValueError:
            print('Enter valid account number.')
            continue
            
        count = 0
        account_transc_list = []
        for i in account_list:
                if i[0]== acc_num:
                    count += 1
                    account_transc_list = i
                    
        if acc_num > 1000 and count==1:
            print('Transaction history')
            print('{}  {} {}'.format(account_transc_list[0],account_transc_list[1],account_transc_list[2]))
            account_transc_list = account_transc_list[4:]
            if len(account_transc_list)%2 == 0:
                for a,b in zip(account_transc_list[::2],account_transc_list[1::2]):
                    print('${:<10}  ${:<10}'.format(a,b))
            else:
                for a,b in zip(account_transc_list[::2],account_transc_list[1::2]):
                    print('${:<10}  ${:<10}'.format(a,b))
                print(account_transc_list[-1])
            break    
        else:
            print('')
            print("Account Number doesn't exist.")
            break
            
# Calls each function by its particular item number. 
switcher = {
        1: create_account,
        2: crediting_and_debiting,
        3: list_account,
        4: transaction_history
    } 
# list of all accounts and transaction.
account_list = []

# Call to main function.
main()
```
